```python
def load_cbs(cbs_data):
    # load cbs data
    cbs = gpd.read_file(cbs_data)
    return cbs

data1 = "../data/cbs/cbs_vk500_2023_v1.gpkg"
```

```python
# check layers in gpkg
layers = gpd.list_layers(data1)
layers

print(f"There is {len(layers)} layer(s) in the gpkg file.")
```
